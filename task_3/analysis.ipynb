### Exploring relations among variables and target

```python
plt.figure(figsize=(8, 12))
heatmap = sns.heatmap(df.corr()[['target']].sort_values(by='target', ascending=False), vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Features Correlating with target', fontdict={'fontsize':18}, pad=16);
```

### Searching for Outliers in dataset

```python
features = [i for i in df.drop("target", axis = 1).columns]

plt.figure(figsize=(10,5))
for i in range(0,len(features)):
    sns.boxplot(y=df[features[i]], color='CornflowerBlue', orient='v')
    plt.show()
```
